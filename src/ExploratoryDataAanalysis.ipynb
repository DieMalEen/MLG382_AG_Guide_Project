```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read from file
df = pd.read_csv('Student_performance_data .csv')
```

```python
#display data
df.head(10)
```

```python
df.info()
```

```python
#separarting categorical data and numerical data
#categorical
categorical_data_columns = df.loc[:,["Gender","Ethnicity","ParentalEducation","Tutoring","Music","Volunteering","Extracurricular"]]
print(categorical_data_columns)
```

```python
#graphs showing the counts of each categorical data column
target = "GradeClass"
plt.figure(figsize=(12, 10))
num_per_row = 3
categorical_features = [col for col in categorical_data_columns if col != target]
rows = (len(categorical_features) + num_per_row - 1) 

fig, axes = plt.subplots(rows, num_per_row, figsize=(num_per_row*6, rows*4))
axes = axes.flatten()

for i, col in enumerate(categorical_features):
    sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f"{col} ")
    axes[i].tick_params(axis='x', rotation=0)
for j in range(i+1, len(axes)):
     fig.delaxes(axes[j])
plt.tight_layout()
plt.suptitle("Univariate Analysis: Categorical Features", fontsize=16, y=1.02)
plt.show()
```

```python
#numerical
numerical_data_columns = df.loc[:,["Age","Absences","StudyTimeWeekly","GPA"]]
print(numerical_data_columns)
```

```python
target = "GradeClass"
num_per_row = 3
num_feats = [col for col in numerical_data_columns if col != target] #why is thier a gap between the ages
rows = (len(num_feats) + num_per_row - 1)
fig, axes = plt.subplots(rows, num_per_row, figsize=(num_per_row*6, rows*4))
axes = axes.flatten()
for i, col in enumerate(num_feats):
    sns.histplot(df[col].dropna(), kde=True, ax=axes[i])
    axes[i].set_title(f"{col} Distribution")
for j in range(i+1, len(axes)):
     fig.delaxes(axes[j])
plt.tight_layout(pad = 2.0)
plt.suptitle("Univariate Analysis: Numerical Features", fontsize=16, y=1.02)
plt.show()
```

```python

num_per_row = 3
rows = (len(num_feats) + num_per_row - 1) // num_per_row

fig, axes = plt.subplots(rows, num_per_row, figsize=(num_per_row*6, rows*4))
axes = axes.flatten()

for i, col in enumerate(num_feats):
    sns.boxplot(x=target, y=col, data=df, ax=axes[i])
    axes[i].set_title(f"{col} by {target}")

for j in range(i+1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.suptitle("Bivariate Analysis: Numerical Features vs Target", fontsize=16, y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
num_per_row = 2
rows = (len(categorical_features) + num_per_row - 1) // num_per_row
fig, axes = plt.subplots(rows, num_per_row, figsize=(num_per_row*8, rows*5))
axes = axes.flatten()
for i, col in enumerate(categorical_data_columns):
    ct = pd.crosstab(df[col], df[target])
    ct.plot(kind='bar', stacked=True, ax=axes[i], colormap="viridis")
    axes[i].set_title(f"{col} vs {target}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=0)
for j in range(i+1, len(axes)):
     fig.delaxes(axes[j])

plt.tight_layout(pad=3.0)
plt.suptitle("Bivariate Analysis: Categorical Features vs Target", fontsize=16, y=1.02)
plt.show()
```
